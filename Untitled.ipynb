```python
import tensorflow as tf
import os
from PIL import Image
import numpy as np
import h5py
from matplotlib import pyplot as plt

%matplotlib inline 
```

```python
def label(filename):
  if 'virus' in filename  or 'bacteria' in filename:
    return 1
  else:
    return 0


def image2vector(path, filename, num_px = 128):
  filename = path + filename
  new_image = Image.open(filename).convert('RGB').resize((num_px,num_px))
  new_image = np.array(new_image)
  return new_image

def load_data(path, num_px):
  X = []
  Y = []
  
  for entry in os.scandir(path):
    if entry.is_file():
       X.append(image2vector(path, entry.name, num_px))
       Y.append([label(entry.name)])

  X = np.array(X)
  Y = np.array(Y)
  
  return X, Y
```

```python
def model (x, y, action):
    
    class myCallback(tf.keras.callbacks.Callback):
      def on_epoch_end(self, epoch, logs={}):
        if(logs.get('accuracy') > 0.98):
          print("\nReached 95% accuracy so cancelling training!")
          self.model.stop_training = True
            
    callbacks = myCallback()

    x, y = x, y
    x = x / 255.0
      
    model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(20, activation=tf.nn.relu),
        tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)
    ])

    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    if (action == 'train'):
        history = model.fit(x, y, epochs = 12, callbacks=[callbacks])
    elif (action == 'evaluate'):
        history = model.fit(x, y, epochs = 12, callbacks=[callbacks])
    elif (action == 'predict'):
        history = model.predict(x)


    #model.evaluate(test_images, test_labels)
    #classifications = model.predict(test_images)
    #print(classifications[0])
    #print(test_labels[0])
    return history
```

```python
path_train = 'D:/Documentos/deep_learning/trabajos/Neumonia/chestxray/train/'
path_dev = 'D:/Documentos/deep_learning/trabajos/Neumonia/chestxray/test/'
path_test = 'D:/Documentos/deep_learning/trabajos/Neumonia/chestxray/val/'
```

```python
x_train, y_train = load_data(path_train, 128)
```

```python
x_dev, y_dev = load_data(path_dev, 128)
```

```python
x_test, y_test = load_data(path_test, 128)
```

```python
# Example of a picture
index = 0
plt.imshow(x_train[index])
print ("y = " + str(np.squeeze(y_train[index])))
```

```python
history = model(x_train, y_train, 'train')
```

```python
history1 = model(x_dev, y_dev, 'evaluate')
```

```python
# evaluacin con img propias o de val
index = 0
plt.imshow(x_test[index])
history3 = model(x_test[index], y_test[index], 'predict')
print ("y = " + str(np.squeeze(y_test[index])))
print ("prediccion = " + str((history3)))
```

```python
epochs = history.epoch
epoch1 = history1.epoch
plt.plot(epochs, history.history['accuracy'], 'g', label='Training accuracy')
plt.plot(epoch1, history1.history['accuracy'], 'r', label='Dev accuracy')

plt.title('Training Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.plot(epochs, history.history['loss'], 'g', label='Training loss')
plt.plot(epoch1, history1.history['loss'], 'r', label='Dev loss')

plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
